# Some Cool Stuff Python can do

In astronomy, observed data is typically stored in fits files, that can be plotted here using the same packages to plot data points. Here, I am going to give an example using a published image of a popular disk, that recently garnered attention for a newly discovered protoplanet in the system, AB Aur b

```python
#import astronomy package
from astropy.io import fits
```

```python
# Load the image.

# Remember to update the image directory
# according to your local file location.
file_dir = 'data/'

# Specify the name of the FITS image file.
img_name = 'pdi_pi_collapsed.fits'

# Read the image data from extension 1 of the FITS file.
img = fits.getdata(file_dir+img_name,ext=1)
```

```python
# Create a figure.
fig = plt.figure(figsize=(8,8))

# Display the image on the figure.
image = plt.imshow(img, origin='lower', vmin=-0.001, vmax=0.01)

# Add a colorbar to the image.
cbar = fig.colorbar(image, shrink=0.82)

# Set the label for the colorbar.
cbar.set_label('Surface Brightness (arb. units)', rotation=90, fontsize=12)

# Set the x-axis display limits.
plt.xlim([60,140])

# Set the y-axis display limits.
plt.ylim([60,140])

# Add the x-axis label.
plt.xlabel(r'$\Delta$ RA [px]', fontsize=12)

# Add the y-axis label.
plt.ylabel(r'$\Delta$ Dec [px]', fontsize=12);
```

Python is very powerful and can read many many rows of data in a short time. In the above example, the image is a 201 x 201 matrix, so a total of 40401 data points. Now let's try reading in 20,000 rows of data with 97 different columns, and use two of those columns to create a scatter plot.

Here, we are loading positional data of 20,000 stars from a CSV table that consists of Gaia (ESA's space telescope) data for the closest stars. We are then showing all the stars as per their positional coordinates RA and Dec (Right Ascension and Declination) as follows

```python
# Import the pandas library for data analysis and table operations.
import pandas as pd
```

```python
# Read the CSV file containing 97 columns of data
# for the closest 20,000 stars from the Gaia Archive (DR2).
stellar = pd.read_csv(file_dir+'closest20kstars.csv')

# Create a matplotlib figure.
fig = plt.figure(figsize=[10,10])

# Plot a scatter graph using RA on the x-axis and Dec on the y-axis.
# alpha: transparency, s: marker size,
# lw: edge line width, ec: edge color.
plt.scatter(stellar['ra'], stellar['dec'], alpha=0.8, s=12, lw=0.5, ec='k', color='darkorchid')

# Add axis limits, labels, and a title to the figure.
plt.xlim([0,362])
plt.ylim([-75,75])
plt.xlabel('RA [$\degree$]', fontsize=14)
plt.ylabel('Dec [$\degree$]', fontsize=14)
plt.savefig("gaia_archive.png")
plt.title('Closest Stars 20,000 stars from Gaia Archive (DR2)', fontsize=16);
```

### Print the Fibonacci sequence

```python
# Initialize the Fibonacci sequence with the first two numbers.
fib = [0, 1]

# Append the next 8 Fibonacci numbers to the list.
for _ in range(8):
    # Add the sum of the last two elements to the end of the list.
    fib.append(fib[-1] + fib[-2])

# Print the complete Fibonacci sequence.
print(fib)
```

```python
# Define a function to print the first n Fibonacci numbers.
def fibonacci(n):
    # Initialize the first two Fibonacci numbers.
    a, b = 0, 1

    # Loop n times.
    for _ in range(n):
        # Print the current Fibonacci number.
        print(a)

        # Update the values for the next Fibonacci number.
        a, b = b, a + b

# Call the function to print the first 10 Fibonacci numbers.
fibonacci(10)
```

```python
# Define a recursive function to calculate the Fibonacci number.
def fib(n):
    # Return n directly if it is 0 or 1.
    if n < 2:
        return n

    # Calculate the Fibonacci number as the sum of the two previous numbers.
    return fib(n-1) + fib(n-2)

# Calculate and print the first 10 Fibonacci numbers.
for i in range(10):
    print(fib(i))
```

```python
# Define a generator function that produces Fibonacci numbers indefinitely.
def fibonacci():
    # Initialize the first two Fibonacci numbers.
    a, b = 0, 1

    # Continuously generate Fibonacci numbers.
    while True:
        # Yield the current Fibonacci number.
        yield a

        # Update the values for the next Fibonacci number.
        a, b = b, a + b

# Create the generator object.
fib = fibonacci()

# Retrieve and print the first 10 Fibonacci numbers from the generator.
for _ in range(10):
    print(next(fib))
```
